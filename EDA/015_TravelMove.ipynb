```python
move['MVMN_CD_1'] = move['MVMN_CD_1'].map(mov_to_nvnm)
move['MVMN_CD_2'] = move['MVMN_CD_2'].map(mov_to_nvnm)
```

## 이동 방식 (운전자/동승자 여부, 공공성, 거리, 자동/비자동) 기준 병합처리
---

### **1번 그룹 (자가용류) → NVMN 1**

* `MOV 1`: 자가용
* `MOV 2`: 렌터카
* `MOV 3`: 캠핑카

- ✅ 개인이 운전하거나 소유/렌트하는 차량류

---

### **2번 그룹 (비개인 운전자형 차량) → NVMN 2**

* `MOV 4`: 택시
  
- ✅ 운전 기사가 있는 개인 교통수단 (유료 대중 운송)

---

### **3번 그룹 (대중교통 일반) → NVMN 3**

* `MOV 5`: 지하철
* `MOV 6`: 고속전철 (ITX 등)
* `MOV 11`: 관광버스
* `MOV 12`: 시외/고속버스
* `MOV 13`: 시내/마을버스
* `MOV 50`: 버스+지하철
  
- ✅ **육상 대중교통**, 가까운 거리나 시외 이동 포함

---

### **4번 그룹 (장거리 교통수단) → NVMN 4**

* `MOV 7`: KTX/SRT
* `MOV 8`: 새마을/무궁화
* `MOV 9`: 항공기
* `MOV 10`: 배/선박
  
- ✅ **장거리 이동을 위한 교통수단**, 고속 or 해상/항공 포함

---

### **5번 그룹 (개인 비자동 교통) → NVMN 5**

* `MOV 14`: 자전거
* `MOV 15`: 도보
  
- ✅ 엔진 없는 **비자동 개인 이동수단**

---

### **6번 그룹 (기타) → NVMN 6**

* `MOV 16`: 기타
  
- ✅ 위의 모든 코드에 해당되지 않는 교통수단

---
