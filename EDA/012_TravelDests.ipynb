### 방문 장소가 집인 곳을 모두 제거 (EDA시, 발생했던 아래의 문구 수정)

- 역, 주유소, 휴게소 - **주유소와 역은 이동수단과 관련된 정보임**
    - 숙소 정보 - **숙소 추천은 나중에 따로**


> 추가) 역과 숙소는 제거하지 않고 추가
- 전주역이나 수원역 같은 경우, 주변 상권이나 내부에 위치한 팝업 스토어 등 역 자체가 관광지 역할을 수행하는 경우가 많기에 지우는 것이 적합하지 않다고 판단.
- 숙소 같은 경우도, 동선 상에 숙소로 이동하는 동선이 필요한 경우가 고려되어야 하기에, 별도로 제거하지 않음

> 수정) 제거를 안하니까, 자꾸 대전역이나 부산역을 추천해줌
- 대전역 같은 경우, 성심당과 같은 주요 동선이 포함되지만, 부산역, 동대구역같은 경우는 필요 없다고 판단
- 필터링 코드를 전체적으로 모두 수정

```python
df = df[~df['VISIT_AREA_NM'].str.contains('집|주유소|나들목|출입로|터미널|역$', na=False)]
```

```python
pattern = (
    r'집|주유소|나들목|출입로|터미널|정류장|환승센터|주차장|'
    r'호텔|모텔|게스트하우스'
    r'\(주\)|사무실|'
    r'카드|은행|보험|증권|'
    r'^\d+$|^\d{2,}번|^[A-Z0-9]{2,}$'
)

df_filtered = df[~df['VISIT_AREA_NM'].str.contains(pattern, case=False, na=False)]
```
