# 데이터 불러오기

```python
import pandas as pd

data_path = '../data/VL_csv/'
travel = pd.read_csv(data_path + 'tn_travel_여행_E.csv')
precost = pd.read_csv(data_path + 'tn_adv_consume_his_사전소비내역_E.csv')
lodgout = pd.read_csv(data_path + 'tn_lodge_consume_his_숙박소비내역_E.csv')
activity = pd.read_csv(data_path + 'tn_activity_consume_his_활동소비내역_E.csv')
```

# 사전 소비 총합 계산

```python
precost.head(2).T
```

```python
precost_per_travel = precost.groupby(['TRAVEL_ID', 'ADV_NM'])['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
precost_per_travel
```

```python
precost_per_travel['ADV_NM'].value_counts()[:5]
```

- 소비 이력이 너무 난잡해서 별도로 활용하지 않음
- 해당 소비내역을 다 합쳐서 사전 소비내역 컬럼에 추가

```python
precost_per_travel = precost.groupby('TRAVEL_ID')['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
precost_per_travel
```

```python
travel = pd.merge(travel, precost_per_travel, on='TRAVEL_ID', how='left').rename({'PAYMENT_AMT_WON': 'PRE_COST'}, axis=1)
```

```python
travel.head(3).T
```

# 숙소 소비 총합 계산

```python
lodgout.head(2).T
```

```python
lodgout_per_travel = lodgout.groupby('TRAVEL_ID')['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
lodgout_per_travel
```

```python
travel = pd.merge(travel, precost_per_travel, on='TRAVEL_ID', how='left').rename({'PAYMENT_AMT_WON': 'LODGOUT_COST'}, axis=1)
```

```python
travel.head(5).T
```

# 활동 소비 총합 계산

```python
activity.head(2).T
```

```python
activity_per_travel = activity.groupby('TRAVEL_ID')['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
activity_per_travel
```

```python
travel = pd.merge(travel, activity_per_travel, on='TRAVEL_ID', how='left').rename({'PAYMENT_AMT_WON': 'ACTIVITY_COST'}, axis=1)
```

```python
travel.fillna({'PRE_COST': 0, 'LODGOUT_COST': 0, 'ACTIVITY_COST': 0}, inplace=True)
```

```python
travel['TOTAL_COST'] = travel['PRE_COST'] + travel['LODGOUT_COST'] + travel['ACTIVITY_COST']

travel['PRE_COST'] = travel['PRE_COST'].astype(int)
travel['LODGOUT_COST'] = travel['LODGOUT_COST'].astype(int)
travel['ACTIVITY_COST'] = travel['ACTIVITY_COST'].astype(int)
```

```python
travel.head(10).T
```

```python
travel.to_csv(data_path + 'tn_travel_여행_E_COST.csv', index=False, encoding='utf-8-sig')
```
