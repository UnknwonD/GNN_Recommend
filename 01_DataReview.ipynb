## 이미지 데이터 탐색

````python
import os
import matplotlib.pyplot as plt
from PIL import Image

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

def visualize_images(img_path):
    valid_exts = ('.png', '.jpg', '.jpeg', '.bmp', '.tiff')
    img_files = [f for f in os.listdir(img_path) if f.lower().endswith(valid_exts)]
    img_files = img_files[:9] # 9개 이미지 시각화

    plt.figure(figsize=(10, 10))
    for i, img_file in enumerate(img_files):
        img = Image.open(os.path.join(img_path, img_file))
        plt.subplot(3, 3, i + 1)
        plt.imshow(img)
        plt.title(img_file, fontsize=8)
        plt.axis('off')

    plt.tight_layout()
    plt.show()


visualize_images(img_path)
````

````python
img_csv = pd.read_csv(os.path.join(label_path, 'tn_tour_photo_관광사진_E.csv'), encoding='utf-8')
img_csv[img_csv['TOUR_PHOTO_SEQ'] == 2].head(5).T
````
