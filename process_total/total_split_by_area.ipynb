```python
import pandas as pd
```

```python
file_list = ['관광사진_fin.csv', '방문지_fin.csv', '여행_fin.csv', '여행객_fin.csv', '이동내역_fin.csv']

region_dict = {
    'e' : '수도권',
    'f' : '동부권',
    'g' : '서부권',
    'h' : '제주도 및 도서지역'
}
```

```python
for file in file_list:
    print(file, ' 진행 중')
    cur = pd.read_csv("./merged_csv/fin/" + file)
    
    if 'TRAVEL_ID' in cur.columns:
        cur['REGION'] = cur['TRAVEL_ID'].str[0]
    else:
        cur['REGION'] = cur['TRAVELER_ID'].str[0]
        
        
    for key, value in region_dict.items():
        target_dir = f"merged_csv/merged_csv_region/{value}/fin/"
        
        cur_region = cur[cur['REGION'] == key]
        
        print(f"지역 : {key}, 크기 : {len(cur_region)}")
        cur_region.to_csv(target_dir + file, index=False)
```
