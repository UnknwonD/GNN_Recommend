```python
import pandas as pd

data_path = './merged_csv/'
travel = pd.read_csv(data_path + '여행_total.csv')
precost = pd.read_csv(data_path + '사전소비내역_total.csv')
lodgout = pd.read_csv(data_path + '숙박소비내역_total.csv')
activity = pd.read_csv(data_path + '활동소비내역_total.csv')
```

# 소비내역
## 사전 소비 총합 계산

```python
precost.head(2).T
```

```python
precost_per_travel = precost.groupby(['TRAVEL_ID', 'ADV_NM'])['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
```

```python
precost_per_travel = precost.groupby('TRAVEL_ID')['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
```

```python
travel = pd.merge(travel, precost_per_travel, on='TRAVEL_ID', how='left').rename({'PAYMENT_AMT_WON': 'PRE_COST'}, axis=1)
```

## 숙소 소비 총합 계산

```python
lodgout_per_travel = lodgout.groupby('TRAVEL_ID')['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
travel = pd.merge(travel, precost_per_travel, on='TRAVEL_ID', how='left').rename({'PAYMENT_AMT_WON': 'LODGOUT_COST'}, axis=1)
```

## 활동 소비 총합 계산

```python
activity_per_travel = activity.groupby('TRAVEL_ID')['PAYMENT_AMT_WON'].sum().reset_index().sort_values('TRAVEL_ID')
travel = pd.merge(travel, activity_per_travel, on='TRAVEL_ID', how='left').rename({'PAYMENT_AMT_WON': 'ACTIVITY_COST'}, axis=1)
```

```python
travel.fillna({'PRE_COST': 0, 'LODGOUT_COST': 0, 'ACTIVITY_COST': 0}, inplace=True)
```

```python
travel['TOTAL_COST'] = travel['PRE_COST'] + travel['LODGOUT_COST'] + travel['ACTIVITY_COST']

travel['PRE_COST'] = travel['PRE_COST'].astype(int)
travel['LODGOUT_COST'] = travel['LODGOUT_COST'].astype(int)
travel['ACTIVITY_COST'] = travel['ACTIVITY_COST'].astype(int)
travel['TOTAL_COST'] = travel['TOTAL_COST'].astype(int)
```

```python
group_map = {
    1: ['1'],
    2: ['3', '6', '9', '12', '13'],
    3: ['2', '11'],
    4: ['4', '7', '23'],
    5: ['5', '8', '28'],
    6: ['10', '21'],
    7: ['22', '26'],
    8: ['24'],
    9: ['27']
}

code_to_group = {}
for group, codes in group_map.items():
    for code in codes:
        code_to_group[code] = group
code_to_group
```

```python
import ast

def map_purpose_group(purpose_str_list):
    return list(set([code_to_group.get(str(code)) for code in purpose_str_list]))
```

```python
import ast

def map_purpose_group(purpose_list):
    group_list = []
    for code in purpose_list:
        str_code = str(code)
        if str_code == '25':  # 25는 반려동물로 따로 추출함
            continue
        group = code_to_group.get(str_code)
        if group is None:
            print(f"알 수 없는 코드: {str_code} (purpose_list: {purpose_list})")
        else:
            group_list.append(group)
    return list(set(group_list))
```

```python
travel_df['TRAVEL_PURPOSE_GROUP'] = travel_df['TRAVEL_PURPOSE'].apply(map_purpose_group)
```
