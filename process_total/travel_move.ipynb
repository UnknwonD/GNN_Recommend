```python
import pandas as pd
```

```python
data_path = './merged_csv/merged_csv_region/제주도 및 도서지역/'
move = pd.read_csv(data_path + "이동소비내역_total.csv")
```

```python
mov_to_nvnm = {
    1: 1,  # 자가용
    2: 1,  # 렌터카
    3: 1,  # 캠핑카
    4: 2,  # 택시
    5: 3,  # 지하철
    6: 3,  # 고속전철
    7: 4,  # KTX/SRT
    8: 4,  # 새마을/무궁화
    9: 4,  # 항공기
    10: 4, # 배
    11: 3, # 관광버스
    12: 3, # 시외버스
    13: 3, # 시내버스
    14: 5, # 자전거
    15: 5, # 도보
    16: 6, # 기타
    50: 3  # 버스+지하철
}
```

```python
move['MVMN_CD_1'] = move['MVMN_CD_1'].map(mov_to_nvnm).fillna(0)
move['MVMN_CD_2'] = move['MVMN_CD_2'].map(mov_to_nvnm).fillna(0)
```
